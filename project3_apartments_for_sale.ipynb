Нужно научиться определять рыночную стоимость объектов недвижимости. Ваша задача — установить параметры. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма.

Кроме того, необходимо будет проверить, чтобы данные в столбце locality_name были приведены к одному регистру для дальнейшего корректного анализа данных 

## Предобработка данных

### Обработка пропусков

#### Столбец ceiling_height (высота потолков, м)

Начнем анализировать пропуски по порядку столбцов. Начнем со столбца ceiling_height (высота потолков, м). Посмотрим на пропуски в этом столбце

```python
display(len(df[df['ceiling_height'].isna()]))

display(df[df['ceiling_height'].isna()].head(10))
```

Пропусков довольно много. Скорее всего, продавец, заполняя данные, просто не владел этой информацией. Единственный параметр, к которому можно привязать высоту потолков - этажность дома, но здесь не хватает такой важной при купле-продаже квартир характеристики, как год постройки дома. Тогда можно бы было сказать, что квартиры в домах 1950-1960 годов постройки, имеющих 5 этажей - это так называемые "хрущевки", высота потолков здесь довольно низкая. Высота потолков в современных домах на порядок выше, чем в домах, построенных до 2000 года. Но исходя из тех данных, что у нас есть, простозаполним пропуски в данном слолбце средним арифметическим значением

```python
df['ceiling_height'] = df['ceiling_height'].fillna(value=df['ceiling_height'].median())
```

Посмотрим на результат

```python
display(df['ceiling_height'].value_counts())
display(len(df[df['ceiling_height'].isna()]))
```

Да, пропуски ушли. Но видим, что в данном столбце есть очень "нелогичное" значения 100 м. Также исправлю на медиану

```python
df['ceiling_height'] = df['ceiling_height'].replace(100.00,df['ceiling_height'].median())
display(df['ceiling_height'].value_counts())
```

Все получилось, идем дальше

#### Столбец floors_total (количество этажей в доме)

Здесь пропусков совсем немного (86 строк). Сначала посмотрим на эти строки

```python
display(len(df[df['floors_total'].isna()]))
display(df[df['floors_total'].isna()].head(10))
```

Бросается в глаза, что пропуски в этажности дома допущены в в строках, в которых в предыдущем шаге мы вставляли медианные значения в столбце "высота потолков". Поэтому, скорее всего, эти продавцы просто не очень добросовестно заполняли данные. Но связать этажность дома с каким-либо другим параметром квартиры не получается. Лучше всего оставить эти пропуски. Но в Шаге 3 нашего проекта необходимо будет добавить в таблицу столбец с этажом (первый, последний, другой). Для этого нам необходима общая этажность дома. Поэтому для чистоты эксперимента заменим все пропуски в этом столбце на максимальное значение в столбце "floors". Доускаю некоторую погрешность такой замены. Например, квартира, находящаяся на 9 этаже, вполне может находится на последнем этаже, но может и на "другом". Но так как пропусков в нашем столбце немного, на общую картину это не повлияет.

```python
df['floors_total'] = df['floors_total'].fillna(value=df['floor'].max())
```

```python
display(len(df[df['floors_total'].isna()]))
```

Все, пропуски "ушли".

#### Столбец living_area (жилая площадь) 

```python
display(len(df[df['living_area'].isna()]))
display(df[df['living_area'].isna()].head(20))
```

Пропусков довольно много. Предполагается,что наиболее правильно заполнить этот параметр средним отношением жилой к общей площади.

```python
living_area_avg = df['living_area'].sum()/df['total_area'].sum()
df['living_area'] = df['living_area'].fillna(value=df['total_area']*living_area_avg)
display(len(df[df['living_area'].isna()]))
```

Все, пропуски в данном столбце заполнились средним отношением жилой площади к общей

#### Столбец kitchen_area (площадь кухни)

Здесь все сделаем аналогично столбцу living_area. Для начала сосчитаем пропуски

```python
display(len(df[df['kitchen_area'].isna()]))
```

```python
kitchen_area_avg = df['kitchen_area'].sum()/df['total_area'].sum()
df['kitchen_area'] = df['kitchen_area'].fillna(value=df['total_area']*kitchen_area_avg)
display(len(df[df['kitchen_area'].isna()]))
```

Заполнено, пропусков нет

#### Столбец is_apartment (апартаменты)

Апартаменты - вид коммерческой недвижимости. Как правило, владельцы осведомлены о том, что их недвижимость не является жилой. Поэтому предполагаю, что пропуски в данном столбце означают, что это жилая недвижимость. Поэтому поменяем все пропуски на False

```python
df['is_apartment'] = df['is_apartment'].fillna(False)
```

```python
display(len(df[df['is_apartment'].isna()]))
display(df.head(20))
```

### Добавление столбца "Цена за квадратный метр"

Рассчитаем цену за 1 м² следующим образом: Цена на момент снятия с публикации /  площадь квартиры в квадратных метрах (м²). Добавим этот столбец в конец таблицы.

### День недели, месяц и год публикации объявления

Для того, чтобы добавить эти данные в таблицу, применим методы .weekday (возвращает дни недели по индексам: понедельник - 0, воскресенье - 6), .month, .year соответственно к столбцу first_day_exposition

```python
df['weekday'] = pd.DatetimeIndex(df['first_day_exposition']).weekday
df['month'] = pd.DatetimeIndex(df['first_day_exposition']).month
df['year'] = pd.DatetimeIndex(df['first_day_exposition']).year
display(df.head())
```

### Добавление столбца "Категория этажа"

```python
def floor_category(row):
    floor = row['floor']
    floors_total = row['floors_total']
    if floor==1:
        return 'Первый'
    if floor==floors_total:
        return 'Последний'
    return 'Другой'
df['floors_category'] = df.apply(floor_category, axis=1)
display(df.head())

```

```python
display(df[['floors_total', 'floor', 'floors_category']])
```

Все получилось, столбец с категорией этажа добавлен.

- близости к центру города (коэффициент корреляции 0,35) - чем ближе к центру, тем дороже,
- количества комнат (коэффициент корреляции 0,15), здесь зависимость очень слабая.
Аналогичная зависимость в Санкт-Петербурге:
- этаж, на котором расположена квартира (дороже всех квартиры, расположенные между первым и последним этажами, дешевле - квартиры на первых этажах),
- близости к центру города (коэффициент корреляции 0,33) - чем ближе к центру, тем дороже. Но здесь прослеживается особенность: есть "престижные" районы далеко от центра (27 км), где стоимость выше, чем в центре города (132 тыс.руб./кв м)
- количества комнат (коэффициент корреляции 0,23).
Если рассматривать только центр Санкт-Петербурга, то здесь в основном цена квадратного метра зависит только от этажа и в некоторой степени от количества комнат (коэф.корреляции 0,3).
